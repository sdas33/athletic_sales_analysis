```python
# Import necessary libraries
import pandas as pd
import numpy as np
import os

# Import necessary libraries
import pandas as pd
import numpy as np
import os

# Define file paths for the datasets using the relative path
sales_2020_path = os.path.join('M5_Starter_Code', 'athletic_sales_2020.csv')
sales_2021_path = os.path.join('M5_Starter_Code', 'athletic_sales_2021.csv')

# Load the datasets into DataFrames
sales_2020 = pd.read_csv(sales_2020_path)
sales_2021 = pd.read_csv(sales_2021_path)

# Display the first few rows of each dataset to inspect column names
print("2020 Sales Data:")
print(sales_2020.head())

print("\n2021 Sales Data:")
print(sales_2021.head())

# Combine and Clean the Data
# Combine the two datasets into a single DataFrame
combined_sales = pd.concat([sales_2020, sales_2021], ignore_index=True)

# Reset the index
combined_sales.reset_index(drop=True, inplace=True)

# Check for null values in the combined data
print("\nNull Values in Combined Data:")
print(combined_sales.isnull().sum())

# Convert the 'invoice_date' column to datetime
combined_sales['invoice_date'] = pd.to_datetime(combined_sales['invoice_date'])

# Confirm the data type conversion
print("\nData Types After Conversion:")
print(combined_sales.dtypes)

# ----> Check if 'qty' column exists, if not, adjust the column name accordingly <----
# Print the columns of the combined_sales DataFrame
print("\nColumns in combined_sales:")
print(combined_sales.columns)

# If the 'qty' column doesn't exist, replace 'qty' with the correct column name
# For example, if the column is named 'quantity', change the following line:
# total_products_sold=('qty', 'sum')  to total_products_sold=('quantity', 'sum')

# Determine which Region Sold the Most Products
products_sold_by_region = combined_sales.groupby(['region', 'state', 'city']).agg(
    total_products_sold=('units_sold', 'sum')  # Replace 'units_sold' with another name if needed
).reset_index()

# Sort by total products sold in descending order
products_sold_by_region.sort_values(by='total_products_sold', ascending=False, inplace=True)

print("\nTop 5 Regions by Products Sold:")
print(products_sold_by_region.head(5))

# Determine which Region had the Most Sales
sales_by_region = combined_sales.groupby(['region', 'state', 'city']).agg(
    total_sales=('total_sales', 'sum')
).reset_index()

# Sort by total sales in descending order
sales_by_region.sort_values(by='total_sales', ascending=False, inplace=True)

print("\nTop 5 Regions by Total Sales:")
print(sales_by_region.head(5))

# Determine which Retailer had the Most Sales
sales_by_retailer = combined_sales.groupby(['retailer', 'region', 'state', 'city']).agg(
    total_sales=('total_sales', 'sum')
).reset_index()

# Sort by total sales in descending order
sales_by_retailer.sort_values(by='total_sales', ascending=False, inplace=True)

print("\nTop 5 Retailers by Total Sales:")
print(sales_by_retailer.head(5))

# Determine which Retailer Sold the Most Women's Athletic Footwear
# Filter for Women's Athletic Footwear
womens_footwear_sales = combined_sales[combined_sales['product_category'] == "Women's Athletic Footwear"]

# Group by retailer, region, state, and city
womens_footwear_by_retailer = womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city']).agg(
    total_products_sold=('qty', 'sum')
).reset_index()

# Sort by total products sold in descending order
womens_footwear_by_retailer.sort_values(by='total_products_sold', ascending=False, inplace=True)

print("\nTop 5 Retailers by Women's Athletic Footwear Sales:")
print(womens_footwear_by_retailer.head(5))

# Determine the Day with the Most Women's Athletic Footwear Sales
# Create a pivot table with "invoice_date" as the index
daily_sales = womens_footwear_sales.pivot_table(
    index='invoice_date',
    values='total_sales',
    aggfunc='sum'
)

# Resample to daily bins
daily_sales = daily_sales.resample('D').sum()

# Sort by total sales in descending order
daily_sales.sort_values(by='total_sales', ascending=False, inplace=True)

print("\nTop 10 Days for Women's Athletic Footwear Sales:")
print(daily_sales.head(10))

# Determine the Week with the Most Women's Athletic Footwear Sales
# Resample to weekly bins
weekly_sales = daily_sales.resample('W').sum()

# Sort by total sales in descending order
weekly_sales.sort_values(by='total_sales', ascending=False, inplace=True)

print("\nTop 10 Weeks for Women's Athletic Footwear Sales:")
print(weekly_sales.head(10))
```
